## Task 4b)
Regularization in general leads to lower weights, hence the weaker shade of gray for the bottom picture (lambda = 1). It is really hard to see from the actual picture that the bottom row is less noisy. To make it clear, we increase the contrast in the bottom row by multiplying the weights by 3. See the second row for result. In this picture, it is clear that there are less 'white spots' midts black parts and less 'black spots' among white parts. In other words, it is less noisy. The reason for this is that the model is penalized for having high weights, thus it will, in general, prefer to only activate the weights that have a big impact on the output, and keep lower weights for the rest,
